```python
import pandas as pd

df = pd.read_csv('/content/us-shein-appliances-3987.csv')
```

```python
for column in df.columns:
    if df[column].dtype == 'object':
        # Jika kolom bertipe object, isi dengan mode
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        # Jika kolom bertipe numerik, isi dengan mean
        df[column].fillna(df[column].mean(), inplace=True)
```

```python
checkDuplicated = df.duplicated().sum()
print(checkDuplicated)
```

```python
# Handling duplicate
df = df.drop_duplicates()
```

```python
# Mengecek duplicate setelah di-handle
handle_duplicate = df.duplicated().sum()

print(f"Jumlah data yang duplikat = {handle_duplicate}")
```

```python
# Identify the most frequent value in 'goods-title-link--jump'
most_frequent_goods_title_link_jump = df['goods-title-link--jump'].mode()[0]

# Filter out rows where 'goods-title-link--jump' is the most frequent value
df_filtered = df[df['goods-title-link--jump'] != most_frequent_goods_title_link_jump].copy()

# Display the shape of the filtered dataframe
print(f"Shape of the original dataframe: {df.shape}")
print(f"Shape of the filtered dataframe: {df_filtered.shape}")

# Update the original dataframe with the filtered data
df = df_filtered
```
